```python
import matplotlib.pyplot as plt
import pickle

molecules = [
    "B+",
    "Be",
    "BeH+",
    "BeH2",
    "BH",
    "BH2+",
    "CH+",
    "CH3OH",
    "CH4",
    "CO",
    "CO2",
    "F2",
    "H2O",
    "H2S",
    # "H3+",
    "H3O+",
    # "H4",
    "H6",
    # "HCI",
    "HF",
    "HOOH",
    "LiH",
    "LiOH",
    "Mg",
    "MgH2",
    "N2",
    # "NaCl",
    "NaH",
    "NeH+",
    "NH",
    "NH2-",
    "NH3",
    "NH4+",
    "OH-",
    "SiH4",
]

data = []
for mole in molecules:
    data.append(pickle.load(open(f"../data/cs_op/{mole}_STO-3G_SINGLET_JW.pckl", "rb")))
```

```python
plt.figure(figsize=(20, 5))

for n in ["n_qubits", "n_taper", "n_cs"]:
    plt.bar(range(len(molecules)), [d[n] for d in data], label=n)
plt.xticks(range(len(molecules)), molecules, rotation=90)
plt.legend()
```
